```python
df_new = df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df_new.head()
```

```python
df_new = df_new.drop_duplicates()
df_new
```

```python
df_new.duplicated().sum()
```

```python
df_new.isnull().sum()
```

```python
df_drop=df_new.dropna(subset=['continent'])
df_drop.head()
```

```python
df_drop.isnull().sum()
```

```python
df_fill=df_drop.fillna(0)
df_fill
```

```python
df_fill.dtypes
```

```python
df_fill['date'] = pd.to_datetime(df_fill['date'])
```

```python
df_fill.dtypes
```

```python
df_fill['month'] = df_fill['date'].dt.month
```

```python
df_fill.head()
```

```python
df_fill.dtypes
```

```python
df_fill.groupby('continent').max()
```

```python
df_groupby = df_fill.groupby("continent").max().reset_index()
df_groupby
```

```python
df_groupby['total_deaths_to_total_cases']=df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby.head()
```

```python
plt.figure(figsize=(10,7))
sns.distplot(df_groupby['gdp_per_capita'])
```

```python
sns.jointplot(data=df_groupby , x = 'total_cases'  , y = 'gdp_per_capita', kind='scatter')
```

```python
sns.pairplot(data=df_groupby)
```

```python
sns.catplot(data=df_groupby , x = 'continent' , y = 'total_cases', kind = 'bar')
```
